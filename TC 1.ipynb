```python
import math
```

```python
def f(x):
    # J√° que o objetivo √© encontrar o valor de x para o qual f(x) √© igual a zero,
    # ou seja, onde x^2 - 2 = 0, que nos leva √† raiz quadrada de 2.
    return x**2 - 2

def bissecao(a,b,tolerancia, n_iteracoes):

  # Teorema do valor Intermediario
  if f(a)*f(b) < 0:
      while(math.fabs(b-a)/2 > tolerancia): # quando for menor que o erro, j√° atingimos o valor suficiente que corresponde a ra√≠z
          ponto_medio = (a+b)/2
          n_iteracoes+=1


          if f(ponto_medio) == 0:
              return ponto_medio
              break
          else:
              if f(a)*f(ponto_medio) < 0:
                  b = ponto_medio
                  anteriorB = ponto_medio
              else:
                  a = ponto_medio
                  anteriorA = ponto_medio

      raiz_real = 2**0.5
      # Como mencionado na descri√ß√£o da solu√ß√£o, utilizamos este m√©todo
      # para encontrar o valor aproximado, considerando o ultimo valor encontrado da aproxima√ß√£o com o anterior
      erro_relativo = abs(anteriorB - ponto_medio)/ponto_medio
      if erro_relativo == 0:
        erro_relativo = abs(anteriorA - ponto_medio)/ponto_medio

      return ponto_medio, erro_relativo, n_iteracoes
  else:
      return 0,0,0
```

```python

a = 1
b = 2
tolerancia = 1e-6
ponto_medio, erro_r, n_iteracoes = bissecao(a,b,tolerancia, 0)
if ponto_medio:
  print(f"A aproxima√ß√£o da ra√≠z encontrada foi: {ponto_medio}\nErro relativo calculado:  {erro_r}\nIterado {n_iteracoes} vezes")
else:
  print(f"N√£o existe ra√≠z no intervalo {a} e {b}.")
```

Podemos perceber que ao utilizarmos o m√©todo de Heron, o erro relativo foi menor em rela√ß√£o ao m√©todo da bisse√ß√£o, o que significa que a porcentagem de aproxima√ß√£o foi maior.

Assim, temos como resultado:

O valor aproximado da raiz quadrada de 2 utilizando o metodo de Heron √©: 1.414213562373095
        
         O erro relativo foi de: 1.1277974130712837e-12
         Iterou 4 vezes

```python
def heron_metodo(tolerancia, aesimo, iteracoes):
  prox_aesimo = 0.5 * (aesimo + 2.0 / aesimo)
  erro_relativo = abs(prox_aesimo - aesimo) / aesimo
  if erro_relativo < tolerancia:
    return prox_aesimo, erro_relativo, iteracoes
  return heron_metodo(tolerancia, prox_aesimo, iteracoes+1)

```

```python
valor_aproximado2, erro_relativo, iteracoes = heron_metodo(tolerancia, 1.0, 0)
print(f"O valor aproximado da raiz quadrada de 2 utilizando o metodo de Heron √©: {valor_aproximado2}\n \
        O erro relativo foi de: {erro_relativo}\n \
        Iterou {iteracoes} vezes")
```
